```python
from scipy.optimize import root_scalar
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Ecuacion de prueba
#x**2 + 15*cos(x) = 40
#f(x) = x**2 + 15*cos(x) - 40
#f(x_solucion) = 0
```

```python
#Grafica de la funcion para ver su comportamiento
def f(x):
  return x**2 + 15*np.cos(x) - 40
#p = d/dx de f
def pf(x):
  return 2*x - 15*np.sin(x)
x = np.linspace(start=-10, stop=10, num=100)
plt.plot(x, f(x))
plt.grid()
plt.axhline(y=0,linewidth=2, color='k')
plt.axvline(x=0,linewidth=2, color='k')
plt.show()
```

```python
#definimos la funcion que le entregamos, el metodo, un valor cercano a la solucion, en este caso 5, por que se aprecia en la grafica, y la derivada de la funcion que es necesaria para el metodo
sol = root_scalar(f, method='Newton', x0=5, fprime = pf)
print(f"Metodo de Newton-Raphson:\n\
- Raiz = {sol.root}\n\
- Iteraciones = {sol.iterations}\n\
- Evaluaciones = {sol.function_calls}\n")
```
